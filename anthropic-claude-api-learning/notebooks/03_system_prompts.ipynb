## Basic system prompt

```python
def ask(user_text, system=""):
    kwargs = dict(model=FAST_MODEL, max_tokens=256,
                  messages=[{"role": "user", "content": user_text}])
    if system:
        kwargs["system"] = system
    return client.messages.create(**kwargs).content[0].text

# No system prompt
print(ask("What time is it?"))
```

## SHOULD / SHOULD NOT structure

```python
concise_system = """
You are a concise assistant.
SHOULD: Answer in one sentence maximum.
SHOULD NOT: Explain your reasoning or add caveats.
"""
print(ask("Explain what DNA is.", system=concise_system))
```
