# Notebook 09 — Prompt Engineering Techniques

Role prompting, few-shot examples, chain-of-thought, XML tags.

```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))
from utils.anthropic_client import client, FAST_MODEL

def ask(prompt, system="", max_tokens=300):
    kwargs = dict(model=FAST_MODEL, max_tokens=max_tokens,
                  messages=[{"role": "user", "content": prompt}])
    if system:
        kwargs["system"] = system
    return client.messages.create(**kwargs).content[0].text
```

## Role prompting

```python
role_prompt = "You are a senior Python engineer. Give code examples with every answer."
print(ask("What is a generator?", system=role_prompt))
```

## Few-shot examples

```python
few_shot = """
Classify the sentiment of the review.

Review: The food was amazing!
Sentiment: positive

Review: Service was slow and the food was cold.
Sentiment: negative

Review: {REVIEW}
Sentiment:"""

for review in ["Best coffee I've ever had!", "Meh, nothing special."]:
    reply = ask(few_shot.replace("{REVIEW}", review))
    print(f"{review!r:45} -> {reply.strip()}")
```

## Chain-of-thought

```python
cot = "Think step by step, then give your final answer on the last line starting with 'Answer:'."
problem = "A train travels 120km in 1.5 hours. What is its average speed?"
print(ask(problem, system=cot))
```

## XML tags for structured input

```python
xml_prompt = """
<document>{DOC}</document>
<task>Summarise the document in one sentence.</task>
"""
doc = "Python was created by Guido van Rossum in 1991. It prioritises readability."
print(ask(xml_prompt.replace("{DOC}", doc)))
```
