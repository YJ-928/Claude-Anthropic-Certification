# Notebook 01 — Getting Started with the Anthropic API

Learn to authenticate, make your first call, and inspect the response object.

```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))
from utils.anthropic_client import client, FAST_MODEL
print("Client ready. Model:", FAST_MODEL)
```

## Your first API call

```python
response = client.messages.create(
    model=FAST_MODEL,
    max_tokens=128,
    messages=[{"role": "user", "content": "Say hello in 10 words."}],
)
print(response.content[0].text)
```

## Inspect the response object

```python
print("Model:       ", response.model)
print("Stop reason: ", response.stop_reason)
print("Input tokens:", response.usage.input_tokens)
print("Output tokens:", response.usage.output_tokens)
print("Content type: ", response.content[0].type)
```
