# Notebook 05 — Controlling Output

Using temperature, pre-filling, and stop sequences to shape Claude's responses.

```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))
from utils.anthropic_client import client, FAST_MODEL
```

## Temperature

```python
prompt = "Give me a creative name for a coffee shop."
for temp in [0.0, 0.5, 1.0]:
    r = client.messages.create(
        model=FAST_MODEL, max_tokens=30, temperature=temp,
        messages=[{"role": "user", "content": prompt}],
    )
    print(f"temp={temp}: {r.content[0].text.strip()}")
```

## Pre-filling the assistant turn

```python
r = client.messages.create(
    model=FAST_MODEL, max_tokens=100,
    messages=[
        {"role": "user",      "content": "List three fruits as JSON array."},
        {"role": "assistant", "content": "["},   # pre-fill
    ],
)
print("[" + r.content[0].text)
```

## Stop sequences

```python
r = client.messages.create(
    model=FAST_MODEL, max_tokens=200,
    stop_sequences=["###"],
    messages=[{"role": "user", "content": "Write a poem, then write ### END"}],
)
print(r.content[0].text)
print(f"\nStop reason: {r.stop_reason}")
```
