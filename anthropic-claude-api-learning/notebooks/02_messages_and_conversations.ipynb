Build conversation history manually using the 3-helper pattern.

```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))
from utils.anthropic_client import client, FAST_MODEL
from utils.message_helpers import make_messages, add_user_message, add_assistant_message, chat
```

## The 3-helper pattern

```python
messages = make_messages()
add_user_message(messages, "My name is Jordan.")
reply = chat(messages)
add_assistant_message(messages, reply)
print("Claude:", reply)
```

## Multi-turn memory

```python
add_user_message(messages, "What did I just tell you my name was?")
reply = chat(messages)
add_assistant_message(messages, reply)
print("Claude:", reply)
```

## Inspect history

```python
for turn in messages:
    print(f"{turn['role'].upper():>10}: {turn['content'][:80]}")
```
