# Notebook 04 — Streaming Responses

Print text as it arrives instead of waiting for the full response.

```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))
from utils.anthropic_client import client, FAST_MODEL
```

## Basic streaming

```python
print("Streaming output: ", end="", flush=True)
full = ""
with client.messages.stream(
    model=FAST_MODEL, max_tokens=200,
    messages=[{"role": "user", "content": "List 5 Python best practices, one per line."}],
) as stream:
    for chunk in stream.text_stream:
        print(chunk, end="", flush=True)
        full += chunk
print()
```

Stream a 200-word story. Count the number of chunks received.

```python
chunks = []
with client.messages.stream(
    model=FAST_MODEL, max_tokens=300,
    messages=[{"role": "user", "content": "Write a 200-word story about a robot."}],
) as s:
    for chunk in s.text_stream:
        chunks.append(chunk)
        print(chunk, end="", flush=True)
print(f"\n\nTotal chunks received: {len(chunks)}")
```
